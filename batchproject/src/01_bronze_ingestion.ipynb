```python
import pyspark.sql.functions as f
```

```python
%sql
USE DATABASE batch;
```

```python
product_df = spark.read.parquet("/Volumes/workspace/source/product/")


product_df = product_df.withColumn("updated_time", f.current_timestamp())
```

```python
customer_df = spark.read.parquet("/Volumes/workspace/source/customer/")


customer_df = customer_df.withColumn("updated_time", f.current_timestamp())

display(customer_df)
```

```python
transactions_df = spark.read.parquet("/Volumes/workspace/source/transactions/")


transactions_df = transactions_df.withColumn("updated_time", f.current_timestamp())
```

```python
transactions_df.write.mode("overwrite").format("delta").saveAsTable("transactions_bronzeLayer")
customer_df.write.mode("overwrite").format("delta").saveAsTable("customer_bronzeLayer")
product_df.write.mode("overwrite").format("delta").saveAsTable("product_bronzeLayer")
```
